####  Pegando somente os últimos dados de cada estado, que representam o dia 22/03:

```python
rdf = df.iloc[765:]
```

####  Número de casos suspeitos do RN:

```python
max(rdf[rdf['state'] == 'Rio Grande do Norte']['suspects'])
```

#### Número de mortes do RN:

```python
max(rdf[rdf['state'] == 'Rio Grande do Norte']['deaths'])
```

#### Estado com o maior número de casos suspeitos:

```python
rdf[rdf['suspects'] == rdf['suspects'].max()]
```

#### Estado com o maior número de mortes:

```python
rdf[rdf['deaths'] == rdf['deaths'].max()]
```

####  Estado com maior número de casos confirmados:

```python
rdf[rdf['cases'] == rdf['cases'].max()]
```

####  Estado com o maior número de casos com diagnóstico negativo:

```python
rdf[rdf['refuses'] == rdf['refuses'].max()]
```

####  Estados com nenhum caso confirmado:

```python
rdf[rdf['cases'] == 0]
```

Todos os estados tem casos confirmados

#### Lista ordenada de casos por estado:

```python
rdf.sort_values(by='cases', ascending = False)
```

#### Lista ordenada de mortes por estado:

```python
rdf.sort_values(by='deaths', ascending = False)
```
